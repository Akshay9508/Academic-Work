```python
airline.dropna(inplace=True)
airline.shape
```

```python
cat_col = ['Gender','Customer Type','Type of Travel','Class']
airline_dummy = pd.get_dummies(airline,columns=cat_col,drop_first=True)
```

```python
airline_dummy.shape
```

# Splitting the data

```python
Y = airline_dummy[['satisfaction']]
X = airline_dummy.drop(columns=['satisfaction'])
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)

len(x_train),len(x_test),len(y_train),len(y_test)
```

# Building the ANN

```python
## Step 1 : Initializing the ANN

ann = tf.keras.models.Sequential()

## Step 2 : Adding the input layer and first hidden layer

ann.add(tf.keras.layers.Dense(units=6,activation='relu'))

## Step 3: Adding the second hidden layer

ann.add(tf.keras.layers.Dense(units=6,activation='relu'))

## Step 4: Adding the output layer
    
ann.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))

## Compiling the ANN

ann.compile(optimizer='adam',loss='binary_crossentropy',metrics='accuracy')

## Training the ANN

ann.fit(x_train,y_train,batch_size=32,epochs=100)
```

```python
y_test
```

```python
y_test['Prediction'] = ann.predict(x_test)
```
