## 3. Hypothesis testing

### To determine whether there is a significant difference in the delivery time at the pizza outlets.

#### Steps in Hypothesis testing
    1. State the Null and Alternate Hypothesis.
    2. Fix the confidence level and significance level.
    3. Choose the appropraite hypothesis test
    4. Implement the test.
    5. Provide inference/ conclusion.

### a. State the Hypotheses 

#### Null hypothesis (Ho) -> 
Delivery time in Electronics_city = MG_road = Whitefield
#### Alternate hypothesis (H1) -> 
Delivery time in Electronics_city, MG_road, Whitefield are not equal

### b. Hypothesis testing

#### Given -> 
Significance level = 0.01, Sample sizes are equal

Confidence level = 1 - 0.01 = 0.99 or 99%

Appropraite hypothesis test would be ANOVA test as we have multiple variables for comparison.

```python
# Implementing the ANOVA test
st.f_oneway(dfoutlets['Electronics_city'],dfoutlets['MG_Road'],dfoutlets['Whitefield'])
```

### c. Inference 

Based on the p-value computed above, we can say null is rejected. 

Therefore, to conclude, the Delivery time of the aforementioned three outlets are not the same.
